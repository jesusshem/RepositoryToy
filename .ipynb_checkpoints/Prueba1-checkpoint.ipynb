# Librerías

```python
from sklearn.datasets import load_boston
import pandas as pd
```

# Datos

```python
boston = load_boston()
```

```python
X = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
X.head()
```

Antes que nada observamos los metadatos para entender de qué se tratan los datos:

```python
print(boston.DESCR)
```

Después análizamos los valores númericos de casa variable para entenderlos mejor

```python
X.describe()
```

Y finalizamos observando si hay alguna correlación entre variables **no dicotómicas**:

```python
X[[col for col in X.columns if col != 'CHAS']].corr()
```
